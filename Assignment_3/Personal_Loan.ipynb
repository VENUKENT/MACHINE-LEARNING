```python
#Naive Bayes 
predictor = np.array(train_norm.iloc[:,[1,2]] )
target = np.array(train_norm['Personal Loan'])
X_train,X_test,y_train,y_test = train_test_split(predictor,target,test_size=0.4,random_state=15)
```

```python
 #Initialize the object and training data
Gs = GaussianNB()
Gs.fit(X_train,y_train)
```

```python
#predictions
Gs_y_predict = Gs.predict(X_test)
print(Gs_y_predict)
```

```python
# prediction accuracy
print(accuracy_score(y_test,Gs_y_predict))
```
